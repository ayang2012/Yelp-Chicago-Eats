Lets take this opportunity to practice webscraping using beautiful soup on a table in wikipedia.

Link: https://en.wikipedia.org/wiki/List_of_neighborhoods_in_Chicago

```python
import requests
url = requests.get('https://en.wikipedia.org/wiki/List_of_neighborhoods_in_Chicago').text

from bs4 import BeautifulSoup
soup = BeautifulSoup(url, 'lxml')
print(soup.prettify())

#then locate the table that we want to extract. Using cmd+f, we find a table class as wikitable sortable
```

```python
my_table = soup.find('table', {'class':'wikitable sortable'})
my_table
```

```python
#I don't know how to parse the columns... so just gonna do the links
links = my_table.findAll('a')
links
# len(all_names)
```

```python
neighborhoods = []
for link in links:
    if link.get('title') not in neighborhoods:
        neighborhoods.append(link.get('title'))
                             
neighborhoods               
```
